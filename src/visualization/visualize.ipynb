```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/apple/Desktop/J.E.R.O.T.I..O/JEROTI/src/data/10seconds-2025-11-27_23-51-43.csv")
df.head()
```

```python
df_filled = df.fillna(0)
```

```python
plt.figure(figsize=(20, 10))
plt.bar(df_filled["name"], df_filled["rss"])
plt.xlabel("Process Name")
plt.ylabel("RSS (Memory)")
plt.title("RSS per Process")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.bar(df_filled["name"], df_filled["threads"])
plt.xlabel("Process Name")
plt.ylabel("Thread Count")
plt.title("Threads per Process")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.bar(df_filled["name"], df_filled["cpu"])
plt.xlabel("Process Name")
plt.ylabel("CPU Usage")
plt.title("CPU Usage per Process")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
