```python
import seaborn as sns

# Check the distribution of 'cpu' and 'ram' usage
df[['cpu', 'ram']].hist(bins=30, figsize=(12, 5))

# Count unique process names
process_counts = df['name'].value_counts()
print(process_counts.head())

# Plot top 10 most frequent processes
process_counts.head(10).plot(kind='bar', title='Top 10 Most Frequent Processes')

# Correlation matrix for numeric columns
corr = df.corr(numeric_only=True)
print(corr)

# Visualize correlation matrix
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['occurrence_count'], bins=30, kde=True)
plt.title('Distribution of Occurrence Count')
plt.xlabel('Occurrence Count')
plt.ylabel('Frequency')
plt.show()

top5_names = process_counts.head(5).index
plt.figure(figsize=(12, 5))
sns.boxplot(data=df[df['name'].isin(top5_names)], x='name', y='cpu')
plt.title('CPU Usage by Top 5 Processes')
plt.show()

plt.figure(figsize=(12, 5))
sns.boxplot(data=df[df['name'].isin(top5_names)], x='name', y='ram')
plt.title('RAM Usage by Top 5 Processes')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='cpu', y='ram', alpha=0.3)
plt.title('CPU vs RAM Usage')
plt.xlabel('CPU')
plt.ylabel('RAM')
plt.show()

df['timestamp_parsed'] = pd.to_datetime(df['timestamp_parsed'])
df_time = df.set_index('timestamp_parsed').resample('1H')[['cpu', 'ram']].sum()
df_time.plot(figsize=(14, 6), title='Total CPU and RAM Usage Over Time')
plt.xlabel('Time')
plt.ylabel('Total Usage')
plt.show()
```
